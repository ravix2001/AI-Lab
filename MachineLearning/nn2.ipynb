```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X1 = 1.5
X2 = 0.625
y = 0.5
```

```python
W1 = np.random.randn(1)
W2 = np.random.randn(1)
print(W1)
print(W2)
```

```python
epochs = 20
alpha = 0.1
errorList = []
wList = []

for epoch in range(epochs):
    # Forward Pass
    y_out = X1*W1 + X2*W2

    # Compute the Error
    error = np.square((y - y_out))
    errorList.append(error)

    # Compute the gradient of Error wrt W
    gradient1 = -2 * (y - y_out) * X1
    gradient2 = -2 * (y - y_out) * X2

    #Backpropagate the Error
    W1 = W1 - gradient1*alpha
    W2 = W2 - gradient2*alpha

print(W1, W2)
```

```python
plt.plot(range(epochs), errorList)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Epochs vs Error")
```

```python
def output(X1, X2, W1, W2):
    return X1*W1 + X2*W2

print(output(X1, X2, W1, W2))
```
